```python
# Load the SpaCy model for NER (Named Entity Recognition)
# Import spacy
import spacy

# Load the SpaCy model for NER (Named Entity Recognition)
nlp = spacy.load("en_core_web_trf")  # Transformer-based model for better performance
```

```python
# extracting text from pdf
def extract_text_from_pdf(pdf_path):
    text = extract_text(pdf_path)
    return text
```

```python
# resume parsing function
def parse_resume(text):
    # Process the resume text with spaCy
    doc = nlp(text)

    # Extract entities related to experience and skills
    skills = []
    experiences = []

    # Assuming skills are usually tagged as "SKILL" or similar in the text
    skill_keywords = ['Python', 'Java', 'C++', 'Machine Learning', 'NLP', 'Data Science', 'AI', 'Deep Learning', 'SQL', 'R']

    # Search for specific skill keywords in the resume
    for token in doc:
        if token.text in skill_keywords:
            skills.append(token.text)

    # Extract experience-related entities (we'll look for job titles or related phrases)
    for ent in doc.ents:
        if ent.label_ == 'ORG':  # Organization names could indicate experience
            experiences.append(ent.text)

    return skills, experiences
```

```python
# HuggingFace Transformers for Better Experience Extraction
# Load a pre-trained model for Named Entity Recognition
# Import required module
from transformers import pipeline

# Load a pre-trained model for Named Entity Recognition
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")

# Define the function to extract skills and experiences
def extract_ner_using_bert(text):
    # Get NER results from HuggingFace BERT model
    entities = ner_pipeline(text)

    skills = []
    experiences = []

    for entity in entities:
        if entity['entity'] == 'B-ORG':  # Organizations for work experience
            experiences.append(entity['word'])
        if entity['entity'] == 'B-MISC':  # Skills or other entities like tools
            skills.append(entity['word'])

    return skills, experiences
```

```python
# Sample resume text (you can upload your resume to test)
# Extract skills using keyword matching
def extract_skills_from_text(text):
    known_skills = [
        'Python', 'Java', 'C++', 'Machine Learning', 'NLP',
        'Data Science', 'AI', 'Deep Learning', 'SQL', 'R', 'TensorFlow'
    ]
    return [skill for skill in known_skills if skill.lower() in text.lower()]

# Extract work experiences using BERT NER model (ORG entities)
def extract_experience_using_bert(text):
    ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", grouped_entities=True)
    entities = ner_pipeline(text)

    experiences = []
    for entity in entities:
        if entity['entity_group'] == 'ORG':
            experiences.append(entity['word'])
    return list(set(experiences))  # remove duplicates

# Combine both methods into a resume parser
def parse_resume_from_text(text):
    skills = extract_skills_from_text(text)
    experiences = extract_experience_using_bert(text)
    return skills, experiences

# Sample resume text
resume_text = """
John Doe
Software Engineer with 5 years of experience in Python, Machine Learning, and Data Science.
Worked at Google and Microsoft.
Skills: Python, TensorFlow, SQL, Machine Learning, Deep Learning
"""

# Parse and print results
skills, experiences = parse_resume_from_text(resume_text)
print("Skills:", skills)
print("Experiences:", experiences)
```
